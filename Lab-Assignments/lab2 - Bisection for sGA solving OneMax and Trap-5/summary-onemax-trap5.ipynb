# 1 - Packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy
```

# 2 - One Max

## 2.1 - Load data

```python
mrps_onemax_1x = pd.read_csv('experiments/mrps_onemax_1x.csv', index_col='run_time')
mrps_onemax_ux = pd.read_csv('experiments/mrps_onemax_ux.csv', index_col='run_time')
evaluations_onemax_1x = pd.read_csv('experiments/evaluations_onemax_1x.csv', index_col='run_time')
evaluations_onemax_ux = pd.read_csv('experiments/evaluations_onemax_ux.csv', index_col='run_time')
```

```python
print('================ Minimally required population size for ONEMAX - 1X ==================')
display(mrps_onemax_1x)
print('=================================== Statitis ====================================')
display(mrps_onemax_1x.describe())
```

```python
print('================ Minimally required population size for ONEMAX - UX ==================')
display(mrps_onemax_ux)
print('=================================== Statitis ====================================')
display(mrps_onemax_ux.describe())
```

```python
print('================== Number of evaluation calls for ONEMAX - 1X ===================')
display(evaluations_onemax_1x)
print('=================================== Statitis ====================================')
display(evaluations_onemax_1x.describe())
```

```python
print('================== Number of evaluation calls for ONEMAX - UX ===================')
display(evaluations_onemax_ux)
print('=================================== Statitis ====================================')
display(evaluations_onemax_ux.describe())
```

## 2.2 - Graph

```python
# convert from pandas type to numpy type
mrps_onemax_1x = mrps_onemax_1x.to_numpy()
mrps_onemax_ux = mrps_onemax_ux.to_numpy()
evaluations_onemax_1x = evaluations_onemax_1x.to_numpy()
evaluations_onemax_ux = evaluations_onemax_ux.to_numpy()
```

```python
# compute mean and standard deviation for graphs
problem_sizes = [10, 20, 40, 80, 160]

# mean and standard deviation of onemax - 1x
mrps_onemax_1x_mean = mrps_onemax_1x.mean(axis=0)

mrps_onemax_1x_std = mrps_onemax_1x.std(axis=0)

evaluations_onemax_1x_mean = evaluations_onemax_1x.mean(axis=0)

evaluations_onemax_1x_std = evaluations_onemax_1x.std(axis=0)

# mean and standard deviation of onemax - ux
mrps_onemax_ux_mean = mrps_onemax_ux.mean(axis=0)

mrps_onemax_ux_std = mrps_onemax_ux.std(axis=0)

evaluations_onemax_ux_mean = evaluations_onemax_ux.mean(axis=0)

evaluations_onemax_ux_std = evaluations_onemax_ux.std(axis=0)
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

# plots
ax.errorbar(problem_sizes, mrps_onemax_1x_mean, yerr=mrps_onemax_1x_std, fmt='.', ecolor='r', linewidth=1, capsize=5)
ax.plot(problem_sizes, mrps_onemax_1x_mean, c='b', label='1X')

ax.errorbar(problem_sizes, mrps_onemax_ux_mean, yerr=mrps_onemax_ux_std, fmt='.', ecolor='r', linewidth=1, capsize=5)
ax.plot(problem_sizes, mrps_onemax_ux_mean, c='g', label='UX')

ax.axis([10, 160, 10, 1e6])
ax.grid(visible=True, which='major', axis='both')
ax.set_title('Tng quan Problem size - MRPS i vi OneMax - POPOP')

#set x_axis
ax.set_xlabel('Problem size')
ax.set_xscale('log')
ax.set_xticks(ticks=problem_sizes, labels=problem_sizes)

# set y_axis
ax.set_yscale('log')
ax.set_ylabel('MRPS')
ax.legend()

plt.show()
```

```python
fig.savefig('graphs/OneMax - problem size - MRPS - 1X - UX.png')
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

# plots
ax.errorbar(problem_sizes, evaluations_onemax_1x_mean, yerr=evaluations_onemax_1x_std, fmt='.', ecolor='r', linewidth=1, capsize=5)
ax.plot(problem_sizes, evaluations_onemax_1x_mean, c='b', label='1X')

ax.errorbar(problem_sizes, evaluations_onemax_ux_mean, yerr=evaluations_onemax_ux_std, fmt='.', ecolor='r', linewidth=1, capsize=5)
ax.plot(problem_sizes, evaluations_onemax_ux_mean, c='g', label='UX')

ax.axis([10, 160, 10, 1e6])
ax.grid(visible=True, which='major', axis='both')
ax.set_title('Tng quan problem size - number of evaluations i vi OneMax - POPOP')

#set x_axis
ax.set_xlabel('Problem size')
ax.set_xscale('log')
ax.set_xticks(ticks=problem_sizes, labels=problem_sizes)

# set y_axis
ax.set_yscale('log')
ax.set_ylabel('MRPS')
ax.legend()

plt.show()
```

```python
fig.savefig('graphs/OneMax - problem size - number of evaluations - 1X - UX.png')
```

## T - test

```python
statistic, p_value = scipy.stats.ttest_ind(mrps_onemax_1x , mrps_onemax_ux, axis=0)

for i in range(len(problem_sizes)):
    print(f'Problem size: {problem_sizes[i]} - t-test:\n -- Statistic: {statistic[i]:0.2f}\n -- p_value: {p_value[i]}')
```

```python
statistic, p_value = scipy.stats.ttest_ind(evaluations_onemax_1x , evaluations_onemax_ux, axis=0)

for i in range(len(problem_sizes)):
    print(f'Problem size: {problem_sizes[i]} - t-test:\n -- Statistic: {statistic[i]:0.2f}\n -- p_value: {p_value[i]}')
```
